```python
df=pd.read_csv('testmarks1.csv')
rno=np.array(df['RollNO'])
eds=np.array(df['EDS'])
son=np.array(df['SON'])
foc=np.array(df['FOC'])
phy=np.array(df['PHY'])
plt.subplot(2,2,1)
plt.title("Result analysis of eds")
plt.xlabel("Roll NO")
plt.ylabel("Marks")
plt.plot(rno,eds)
plt.subplot(2,2,2)
plt.title("Result analysis of SON")
plt.xlabel("Roll NO")
plt.ylabel("Marks")
plt.plot(rno,son)
plt.subplot(2,2,3)
plt.title("Result analysis of PHY")
plt.xlabel("Roll NO")
plt.ylabel("Marks")
plt.plot(rno,phy)
plt.subplot(2,2,4)
plt.title("Result analysis of FOC")
plt.xlabel("Roll NO")
plt.ylabel("Marks")
plt.plot(rno,foc)
plt.show()

```
